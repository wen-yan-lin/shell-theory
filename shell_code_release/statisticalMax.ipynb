```python
import numpy as np
from sklearn.metrics import pairwise_distances
import matplotlib.pyplot as plt
```

```python
testFeat = np.load('../STL-10/resNet50.npy')
testFeat = testFeat/np.linalg.norm(testFeat, axis=1, keepdims=True)
```

```python
d = pairwise_distances(testFeat)
d = d.reshape([d.size])
```

```python
plt.rcParams["font.weight"] = "bold"
plt.rcParams["font.size"] = 15
plt.rcParams["axes.labelweight"] = "bold"


plt.hist(d, bins=100, log=True);
plt.axvline(x=1.42, color='black', linewidth=2, label='s', linestyle='--')
plt.axvline(x=2, color='red', linewidth=2, label='m')
plt.xlabel('Distance')
plt.ylabel('Log Frequency')

plt.legend(['Statistical Max', 'Geometric Max'])
plt.savefig('statisticalMax.pdf', bbox_inches='tight')
```
