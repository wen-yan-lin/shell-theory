```python
from keras.preprocessing import image
from keras.applications.vgg16 import VGG16
from keras.applications.vgg16 import preprocess_input
from keras import backend as K
import numpy as np

import dirUtil 
from highDimLearning import vgg2feat

import tensorflow as tf

from sklearn.metrics.pairwise import euclidean_distances

import matplotlib.pyplot as plt



def label(feat, w, sig):

    numClass = len(w)
    score = np.zeros([feat.shape[0], numClass])
    for i in range(numClass):
        ww = w[i]
        ww = np.reshape(ww, [ww.size, 1])
        sigg = np.reshape(sig[i], [sig[i].size, 1])
        score[:, i:i+1] = np.matmul(feat, ww)-sigg
    score = np.absolute(score)
    labels = np.argmin(score, axis=1)
    return labels, score

    

def feat2labels(feat):
    numClass = len(feat)
    numDim = feat[0].shape[1]
    numPts = 0
    for i in range(numClass):
        numPts = numPts + feat[i].shape[0]

    allFeat = np.zeros([numPts, numDim])
    allLabels = np.zeros(numPts, dtype = int)
    cur = 0
    for i in range(numClass):
        allFeat[cur:cur+feat[i].shape[0],:] = feat[i]
        allLabels[cur:cur+feat[i].shape[0]] = i
        cur = cur + feat[i].shape[0]
    return allFeat, allLabels
        

trainingNum = 1000

tf.reset_default_graph()
K.clear_session()
folder = '/media/daniel/D/code/keras-tutorial/animals/dogs/'
feats = dirUtil.dir2vgg16(folder)
feat1 = vgg2feat(feats)


folder2 = '/media/daniel/D/code/keras-tutorial/animals/cats/'
tf.reset_default_graph()
K.clear_session()
feats = dirUtil.dir2vgg16(folder2)
feat2 = vgg2feat(feats)

folder3 = '/media/daniel/D/code/keras-tutorial/animals/panda/'
tf.reset_default_graph()
K.clear_session()
feats = dirUtil.dir2vgg16(folder3)
feat3 = vgg2feat(feats)





allFeat, gt = feat2labels([feat1, feat2])


####################################
from sklearn.cluster import KMeans

numClass = 2
kmeans = KMeans(n_clusters=numClass, random_state=0).fit(allFeat)
means = kmeans.cluster_centers_

means = means/np.linalg.norm(means, axis =1, keepdims=True)
sigInit = np.zeros([numClass])
labels = kmeans.predict(allFeat)
for i in range(numClass):
    mask = labels == i
    sigInit[i] = np.mean(np.matmul(allFeat, np.transpose(means)))


c = np.matmul(allFeat, np.transpose(means)) - sigInit
plt.scatter(c[:feat1.shape[0], 0],
            c[:feat1.shape[0], 1],
            color='red', alpha=0.1)
plt.scatter(c[feat1.shape[0]:, 0],
            c[feat1.shape[0]:, 1],
            color='blue', alpha=0.1)
plt.show();


```

```python

def multiModel(feat,  meanEst, sigEst, weight=0.0, numIter=10000):

    #numInstances = feat.shape[0]
    numDim = meanEst.shape[1]
    #numShells = int(np.max(gtClass)+1)

    f_ = tf.constant(feat, dtype="float32")
    W_ = tf.get_variable("W", dtype=tf.float32,
                         initializer=meanEst.astype("float32"))
    Sig_ = tf.get_variable("S", dtype=tf.float32,
                           initializer=sigEst.astype("float32"))
    lambda_ = tf.constant(weight, dtype="float32")
    numFeat_ = tf.constant(feat.shape[0], dtype="float32")

    #advSig_ = tf.get_variable("AS", [numShells]) # only works for two class

    projDist_ = tf.matmul(f_, W_)

    meanNorm_ = tf.sqrt(tf.reduce_sum(tf.math.square(W_), axis=0))
    regularizer_ = 0.0*tf.math.abs(1-Sig_) + (1-meanNorm_)*(1-meanNorm_)
    regularizer_ = tf.reduce_sum(regularizer_)

    # dist_ = tf.nn.leaky_relu(projDist_-Sig_)
    # selfCost = tf.reduce_max(dist_, axis =1)
    # selfCost = tf.reduce_sum(selfCost)/numFeat_

    dist_ = tf.math.pow(projDist_-Sig_, 2)
    #selfCost = tf.reduce_min(dist_, axis=1)
    #selfCost = tf.reduce_sum(selfCost)/numFeat_

    #k = tf.reduce_min(dist_, axis=1)

    advCost = tf.reduce_max(dist_, axis=1)
    #advCost = tf.reduce_sum(advCost)/numFeat_
    selfCost = tf.reduce_min(dist_, axis=1)
    selfCost = selfCost/advCost
    selfCost = tf.reduce_sum(selfCost)/numFeat_

    final_cost = selfCost + lambda_*regularizer_

    # dist_ = tf.math.pow(projDist_-Sig_, 2)
    # selfCost = binaryEncoding_*dist_
    # selfCost = tf.reduce_sum(selfCost, axis =1)

    # advCost = invBinaryEncoding_*dist_
    # advCost = tf.reduce_sum(advCost, axis=1)

    # rCost = selfCost/advCost
    # rCost = tf.reduce_sum(rCost)/numFeat_

    # final_cost = rCost + lambda_*regularizer_

    optimizer = tf.train.GradientDescentOptimizer(0.01)
    train = optimizer.minimize(final_cost)

    with tf.Session() as sess:
        sess.run(tf.global_variables_initializer())
        #a = sess.run(W_)
        #print(a[0:4,:])
        
        for i in range(numIter):
            sess.run(train)
            #a = sess.run(k)
            #print(k)
            if i%500 == 0:
               
                d = sess.run(final_cost)
                print(i, d)

        W = sess.run(W_)
        Sig = sess.run(Sig_)
        #d = sess.run(catMat)
        #k = donk
    return W, Sig
        
```

```python

def multiModel2(feat,  meanEst, sigEst, initMask, weight=0.0, numIter=10000):

    numDim = meanEst.shape[1]

    f_ = tf.constant(feat, dtype="float32")
    W_ = tf.get_variable("W", dtype=tf.float32,
                         initializer=meanEst.astype("float32"))
    Sig_ = tf.get_variable("S", dtype=tf.float32,
                           initializer=sigEst.astype("float32"))
    lambda_ = tf.constant(weight, dtype="float32")
    numFeat_ = tf.constant(feat.shape[0], dtype="float32")

    
    mask = tf.placeholder("float", feat.shape[0])
    numInMask = tf.placeholder("float", shape=())


    projDist_ = tf.matmul(f_, W_)

    meanNorm_ = tf.sqrt(tf.reduce_sum(tf.math.square(W_), axis=0))
    regularizer_ = 0.0*tf.math.abs(1-Sig_) + (1-meanNorm_)*(1-meanNorm_)
    regularizer_ = tf.reduce_sum(regularizer_)

   
    dist_ = tf.math.pow(projDist_-Sig_, 2)
   

    advCost = tf.reduce_max(dist_, axis=1)
    selfCost = tf.reduce_min(dist_, axis=1)
    selfCost = selfCost/advCost
    selfCost = mask*selfCost
    selfCost = tf.reduce_sum(selfCost)/numInMask

    final_cost = selfCost + lambda_*regularizer_

   

    optimizer = tf.train.GradientDescentOptimizer(0.01)
    train = optimizer.minimize(final_cost)

    with tf.Session() as sess:
        sess.run(tf.global_variables_initializer())
        #a = sess.run(W_)
        #print(a[0:4,:])
        
        curMask = copy.deepcopy(initMask)
        curNum = sum(curMask)
        print(sum(curMask))
        for i in range(5001):
            sess.run(train, feed_dict={mask: curMask, numInMask: curNum})
            a = sess.run(advCost, feed_dict={mask: curMask, numInMask: curNum})
            if i%100 == 0 and i >0:
                W = sess.run(W_)
                Sig = sess.run(Sig_)
                c = np.matmul(allFeat,W) -Sig
                d = np.absolute(c)                
                ns = np.min(d,axis=1)/np.max(d, axis =1)
                thres = np.percentile(ns,50+ i/100.0)
                curMask = ns<thres
                curNum = np.sum(curMask)
                print(i,thres)



        W = sess.run(W_)
        Sig = sess.run(Sig_)
        #d = sess.run(catMat)
        #k = donk
    return W, Sig
        
```
