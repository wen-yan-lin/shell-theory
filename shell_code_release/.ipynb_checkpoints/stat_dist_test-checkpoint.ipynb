```python
from keras.preprocessing import image
from keras.applications.vgg16 import VGG16
from keras.applications.vgg16 import preprocess_input
from keras import backend as K
import numpy as np

import dirUtil 
from highDimLearning import vgg2feat

import tensorflow as tf

from sklearn.metrics.pairwise import euclidean_distances

import matplotlib.pyplot as plt


import os
import random 
    
def feat2labels(feat):
    numClass = len(feat)
    numDim = feat[0].shape[1]
    numPts = 0
    for i in range(numClass):
        numPts = numPts + feat[i].shape[0]

    allFeat = np.zeros([numPts, numDim])
    allLabels = np.zeros(numPts, dtype = int)
    cur = 0
    for i in range(numClass):
        allFeat[cur:cur+feat[i].shape[0],:] = feat[i]
        allLabels[cur:cur+feat[i].shape[0]] = i
        cur = cur + feat[i].shape[0]
    return allFeat, allLabels


def feat2vec(dataSet):
    numDim = dataSet[0].size
    feat = np.zeros([len(dataSet), numDim])

    runner = 0
    for f in dataSet:
    #f = np.reshape(f, [1, 25088, 1])
        feat[runner, :] = np.reshape(f, [25088])
        runner = runner + 1

    #n = np.linalg.norm(feat, axis=1, keepdims=1)
    #feat = feat/n
    return feat


tf.reset_default_graph()
K.clear_session()
folder = '/media/daniel/D/code/keras-tutorial/animals/dogs/'
feats = dirUtil.dir2vgg16(folder)
feat1 = feats


folder2 = '/media/daniel/D/code/keras-tutorial/animals/cats/'
tf.reset_default_graph()
K.clear_session()
feats = dirUtil.dir2vgg16(folder2)
feat2 = feats

folder3 = '/media/daniel/D/code/keras-tutorial/animals/panda/'
tf.reset_default_graph()
K.clear_session()
feats = dirUtil.dir2vgg16(folder3)
feat3 = feats

feat = feat2vec(feat2)
feat_norm = feat/np.linalg.norm(feat, axis =1, keepdims=True)
```

```python
from sklearn.metrics import pairwise_distances

dim = 25088
num_dist = 100
num_data = 1000
step = num_data//num_dist

rand_mean = np.random.normal(size=(num_dist,dim))
rand_var = np.random.normal(size=(num_dist,1))

rand_mat = np.random.normal(size=(num_data,dim))
for i in range(rand_mean.shape[0]):
    rand_mat[i*step:(i+1)*step] = rand_mat[i*step:(i+1)*step]*rand_var[i]+rand_mean[i]
    #rand_mat[i*10:(i+1)*10] = pow(rand_mat[i*10:(i+1)*10],1/i)

norm_rand_mat = rand_mat/np.linalg.norm(rand_mat, axis =1, keepdims=True)
```

```python
from sklearn.metrics import pairwise_distances

dim = 25088
num_data = 1000

num_dist1 = 100
step1 = num_data//num_dist1

rand_mean = np.random.normal(size=(num_dist1,dim))
rand_var = np.random.normal(size=(num_dist1,1))

rand_mat = np.random.normal(size=(num_data,dim))
for i in range(rand_mean.shape[0]):
    rand_mat[i*step1:(i+1)*step1] = rand_mat[i*step1:(i+1)*step1]*rand_var[i]+rand_mean[i]
    #rand_mat[i*10:(i+1)*10] = pow(rand_mat[i*10:(i+1)*10],1/i)

norm_rand_mat = rand_mat/np.linalg.norm(rand_mat, axis =1, keepdims=True)

num_dist2 = 2
step2 = num_data//num_dist2

rand_mean = np.random.normal(size=(num_dist2,dim))
rand_var = np.random.normal(size=(num_dist2,1))

rand_mat = np.random.normal(size=(num_data,dim))
for i in range(rand_mean.shape[0]):
    rand_mat[i*step2:(i+1)*step2] = rand_mat[i*step2:(i+1)*step2]*rand_var[i]+rand_mean[i]
    #rand_mat[i*10:(i+1)*10] = pow(rand_mat[i*10:(i+1)*10],1/i)

norm_rand_mat2 = rand_mat/np.linalg.norm(rand_mat, axis =1, keepdims=True)
```

```python

plt.tight_layout()



d_norm_pairwise = pairwise_distances(norm_rand_mat, norm_rand_mat)
d_norm_pairwise2 = pairwise_distances(norm_rand_mat2, norm_rand_mat2)


fig, axs = plt.subplots(2, 2)
fontsize = 8

i = 0
j = 0
axs[i,j].set_title('Complex (100 Distributions)', fontweight ='bold', fontsize =fontsize)
h = axs[i,j].hist(d_norm_pairwise.flatten(),bins=100, range=[0,4])
axs[i,j].yaxis.set_visible(False) 
plt.sca(axs[i,j])
plt.xticks([0,1.42,4]);
i = 1
j = 0
h = axs[i,j].hist(d_norm_pairwise.flatten(),bins=100, range=[0,4])
axs[i,j].set_yscale('log')
axs[i,j].yaxis.set_visible(False) 
plt.sca(axs[i,j])
plt.xticks([0,1.42,4]);


i = 0
j = 1
axs[i,j].set_title('Simple (2 Distributions)', fontweight ='bold', fontsize =fontsize)
h = axs[i,j].hist(d_norm_pairwise2.flatten(),bins=50, range=[0,4])
axs[i,j].yaxis.set_visible(False) 
plt.sca(axs[i,j])
plt.xticks([0,1.42,4]);
i = 1
j = 1
h = axs[i,j].hist(d_norm_pairwise2.flatten(),bins=100, range=[0,4])
axs[i,j].set_yscale('log')
axs[i,j].yaxis.set_visible(False) 
plt.sca(axs[i,j])
plt.xticks([0,1.42,4]);

axs[0,0].yaxis.set_visible(True) 
axs[0,0].set_ylabel('Histogram', fontweight ='bold', fontsize =fontsize );
axs[1,0].yaxis.set_visible(True) 
axs[1,0].set_ylabel('Log Histogram', fontweight ='bold',  fontsize =fontsize );
fig.suptitle("Range of possible distances",y=0.05, fontweight ='bold',  fontsize =fontsize)
fig.savefig('/media/daniel/D/Data/tmp/simu_tree.pdf');

```
