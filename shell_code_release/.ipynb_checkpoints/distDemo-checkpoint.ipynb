```python
from shellUtil import classAndOutliers
from displayUtil import histogramByGroundTruth

dataMask = classAndOutliers(gtAll, 0, 0.1, outlierClassList=[1])
feat = allFeat[dataMask]
gt = gtAll[dataMask]
```

```python
import scipy.stats as stats
import math
def pltGaussian(mu=0, variance=1, scale=1,oneSide=False):
    sigma = math.sqrt(variance)
    if oneSide:
        x = np.linspace(0, mu + 3*sigma, 100)
    else:
        x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)

    plt.plot(x, scale*stats.norm.pdf(x, mu, sigma), linewidth=2)
```

## Display Distance from Mean

```python
plt.rcParams["font.weight"] = "bold"
plt.rcParams["font.size"] = 15
plt.rcParams["axes.labelweight"] = "bold"


m = np.mean(feat[gt==0], axis=0, keepdims=True)
ff = np.linalg.norm(feat-m, axis=1)
histogramByGroundTruth(ff, gt==0, bins=100)
pltGaussian(0, 0.2, 200, oneSide=True)
pltGaussian(0, 0.1, 200, oneSide=True)
pltGaussian(0, 0.05, 200, oneSide=True)
```
