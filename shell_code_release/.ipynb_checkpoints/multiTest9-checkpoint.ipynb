```python
from keras.preprocessing import image
from keras.applications.vgg16 import VGG16
from keras.applications.vgg16 import preprocess_input
from keras import backend as K
import numpy as np

import dirUtil 
from highDimLearning import vgg2feat

import tensorflow as tf

from sklearn.metrics.pairwise import euclidean_distances

import matplotlib.pyplot as plt




    

def feat2labels(feat):
    numClass = len(feat)
    numDim = feat[0].shape[1]
    numPts = 0
    for i in range(numClass):
        numPts = numPts + feat[i].shape[0]

    allFeat = np.zeros([numPts, numDim])
    allLabels = np.zeros(numPts, dtype = int)
    cur = 0
    for i in range(numClass):
        allFeat[cur:cur+feat[i].shape[0],:] = feat[i]
        allLabels[cur:cur+feat[i].shape[0]] = i
        cur = cur + feat[i].shape[0]
    return allFeat, allLabels
        
    


trainingNum = 1000

tf.reset_default_graph()
K.clear_session()
folder = '/home/sliu/code/keras-tutorial/animals/dogs/'
feats = dirUtil.dir2vgg16(folder)
feat1 = vgg2feat(feats)


folder2 = '/home/sliu/code/keras-tutorial/animals/cats/'
tf.reset_default_graph()
K.clear_session()
feats = dirUtil.dir2vgg16(folder2)
feat2 = vgg2feat(feats)

folder3 = '/home/sliu/code/keras-tutorial/animals/panda/'
tf.reset_default_graph()
K.clear_session()
feats = dirUtil.dir2vgg16(folder3)
feat3 = vgg2feat(feats)

allFeat, gt = feat2labels([feat1, feat2, feat3])
```

```python
import numpy as np
import tensorflow as tf
import copy

from sklearn.cluster import KMeans



def multiModelB(feat,  meanEst, sigEst, sigA, weight=0.0, numIter=10000):

    numDim = meanEst.shape[1]

    f_ = tf.constant(feat, dtype="float32")
    W_ = tf.get_variable("W", dtype=tf.float32,
                         initializer=meanEst.astype("float32"))
    Sig_ = tf.get_variable("S", dtype=tf.float32,
                           initializer=sigEst.astype("float32"))
    Sig_adv = tf.get_variable("Sa", dtype=tf.float32,
                            initializer=sigA.astype("float32"))

    lambda_ = tf.constant(weight, dtype="float32")
    numFeat_ = tf.constant(feat.shape[0], dtype="float32")


   


    projDist_ = tf.matmul(f_, W_)

    meanNorm_ = tf.sqrt(tf.reduce_sum(tf.math.square(W_), axis=0))
    regularizer_ = 0.0*tf.math.abs(1-Sig_) + (1-meanNorm_)*(1-meanNorm_)
    regularizer_ = tf.reduce_sum(regularizer_)

    
    projDiff = projDist_-Sig_
    dist_ = tf.math.pow(projDiff,2)

    
    advCost = tf.reduce_max(projDiff, axis=1)
    #advCost = tf.nn.relu(advCost)     
    advCost = tf.math.pow(advCost, 2)
    selfCost = tf.reduce_sum(dist_, axis=1) 
    selfCost = (selfCost-advCost)/(advCost+0.0001)
    selfCost = tf.reduce_sum(selfCost)/numFeat_
    
    
    
    assign_ = tf.math.softmax(projDiff/0.01)
    dist_2 = 0.01*tf.math.abs(projDiff- Sig_ -Sig_adv)
    #dist_2 = tf.math.pow(projDiff- Sig_ -Sig_adv, 2)
    dCost = dist_2 *assign_
    dCost = tf.reduce_sum(dist_2, axis =1)
    dCost = tf.reduce_sum(dCost)/numFeat_
    
    final_cost = selfCost + lambda_*regularizer_ + 1.0*dCost

    optimizer = tf.train.GradientDescentOptimizer(0.01)
    train = optimizer.minimize(final_cost)

    with tf.Session() as sess:
        sess.run(tf.global_variables_initializer())
        for i in range(numIter):
            sess.run(train)
            if i%500 == 0:
               
                d = sess.run(final_cost)
                print(i, d)


        W = sess.run(W_)
        Sig = sess.run(Sig_)
        SigAdv = sess.run(Sig_adv)
    return W, Sig, SigAdv
        

def multiModel2B(feat,  meanEst, sigEst, sigA, initMask, weight=0.0, numIter=10000):

    numDim = meanEst.shape[1]

    f_ = tf.constant(feat, dtype="float32")
    W_ = tf.get_variable("W", dtype=tf.float32,
                         initializer=meanEst.astype("float32"))
    Sig_ = tf.get_variable("S", dtype=tf.float32,
                           initializer=sigEst.astype("float32"))
    Sig_adv = tf.get_variable("Sa", dtype=tf.float32,
                           initializer=sigA.astype("float32"))

    lambda_ = tf.constant(weight, dtype="float32")
    numFeat_ = tf.constant(feat.shape[0], dtype="float32")

    
    mask = tf.placeholder("float", feat.shape[0])
    numInMask = tf.placeholder("float", shape=())
        
    
    
    
    projDist_ = tf.matmul(f_, W_)

    meanNorm_ = tf.sqrt(tf.reduce_sum(tf.math.square(W_), axis=0))
    regularizer_ = 0.0*tf.math.abs(1-Sig_) + (1-meanNorm_)*(1-meanNorm_)
    regularizer_ = tf.reduce_sum(regularizer_)

    
    projDiff = projDist_-Sig_
    dist_ = tf.math.pow(projDiff,2)

    
    advCost = tf.reduce_max(projDiff, axis=1)
    advCost = tf.nn.relu(advCost)     
    advCost = tf.math.pow(advCost, 2)
    selfCost = tf.reduce_sum(dist_, axis=1) 
    selfCost = (selfCost-advCost)/(advCost+0.0001)
    selfCost = mask*selfCost
    selfCost = tf.reduce_sum(selfCost)/numInMask
    
    
    
    assign_ = tf.math.softmax(projDiff/0.01)
    dist_2 = 0.01*tf.math.abs(projDiff- Sig_ -Sig_adv)
    #dist_2 = tf.math.pow(projDiff- Sig_ -Sig_adv, 2)
    dCost = dist_2 *assign_
    dCost = tf.reduce_sum(dist_2, axis =1)
    dCost = mask*dCost
    dCost = tf.reduce_sum(dCost)/numInMask
    
    final_cost = selfCost + lambda_*regularizer_ + 1.0*dCost
    
    
    

    optimizer = tf.train.GradientDescentOptimizer(0.01)
    train = optimizer.minimize(final_cost)

    with tf.Session() as sess:
        sess.run(tf.global_variables_initializer())
        
        curMask = copy.deepcopy(initMask)
        curNum = sum(curMask)
        for i in range(5001):
            sess.run(train, feed_dict={mask: curMask, numInMask: curNum})
            a = sess.run(advCost, feed_dict={mask: curMask, numInMask: curNum})
            if i%100 == 0 and i >0:
                W = sess.run(W_)
                Sig = sess.run(Sig_)
                c = np.matmul(feat,W) -Sig
                ns = np.max(c,axis=1)
                thres = np.percentile(ns,50- i/100.0)
                curMask = ns>thres
                #curMask = np.logical_and(ns>thres, ns>0.01)
                curNum = np.sum(curMask)
                print(i,thres)

        for i in range(10000):
            sess.run(train, feed_dict={mask: curMask, numInMask: curNum})

        W = sess.run(W_)
        Sig = sess.run(Sig_)
        SigAdv = sess.run(Sig_adv)

    return W, Sig, SigAdv
        



def clusterData(allFeat, numClass=2):

    tf.reset_default_graph()
    
    #embed = umap.UMAP().fit(allFeat)
    #f = embed.transform(allFeat)
    
    f = allFeat
    kmeans = KMeans(n_clusters=numClass, random_state=0).fit(f)
    labelK = kmeans.predict(f)
    means =np.zeros([numClass, allFeat.shape[1]])

    for i in range(numClass):
        mask = labelK == i
        means[i,:] = np.mean(allFeat[mask,:], axis =0)
        
    means = means/np.linalg.norm(means, axis =1, keepdims=True)
    sigInit = np.zeros([numClass])
    for i in range(numClass):
        mask = np.logical_not(labelK == i)
        sigInit[i] = np.mean(np.matmul(allFeat[mask,:], np.transpose(means[i,:])))

    score = np.min(kmeans.transform(f), axis =1)    
    val = np.percentile(score, 50)
    maskK = score<val


    W = copy.deepcopy(np.transpose(means))
    Sig = copy.deepcopy(np.transpose(sigInit))
        

    
    W, Sig, SigA = multiModelB(allFeat[maskK,:], W, Sig, Sig,
                        weight=1.0, numIter=10000)
    c = np.matmul(allFeat,W) -Sig
    label = np.argmax(c, axis =1)
    #return W, Sig, SigA, label, c, labelK 


    
    tf.reset_default_graph()
    W, Sig, SigA = multiModel2B(allFeat, W, Sig, SigA, maskK, weight=1.0, numIter=10000)
    
    #tf.reset_default_graph()
    #W, Sig, SigA = multiModelB(allFeat, W, Sig, SigA,
    #                  weight=1.0, numIter=10000)

    c = np.matmul(allFeat,W) -Sig
    label = np.argmax(c, axis =1)
    return W, Sig, SigA, label, c, labelK 
```

```python
from sklearn.metrics import pairwise_distances
from sklearn.preprocessing import StandardScaler

from ClusteringTree import search_k_nearest_neighbour
from ClusteringTree import evaluate_clustering

def clustering_new_predict(feats_test, W, Sig):
    
    test_response = np.matmul(feats_test, W) - Sig
    test_score = np.max(test_response, axis=1)
    test_label = np.argmax(test_response, axis=1)
    
    return test_label, test_score

def evaluate_clustering(test_label, test_score, train_score, lab, NN_table, val):
    
    n_neighbors = NN_table.shape[1]
    count_arr = np.zeros(n_neighbors)
    drop_arr = np.zeros(n_neighbors)
    for i in range(test_label.shape[0]):
        knn_indces = NN_table[i, :]
        for j in range(n_neighbors):
            ind = knn_indces[j]
            if test_score[i] < val: # and train_score[ind] < val:
                count_arr[j] += 1
                drop_arr[j] += 1
            elif test_label[i] == lab[ind]:
                count_arr[j] += 1

    for i in range(n_neighbors):
        print(i+1,"-th NN: ", count_arr[i], "/", test_label.shape[0], " - ", drop_arr[i])
    print("Overall: ", np.sum(count_arr)*100/(test_label.shape[0]*n_neighbors),"%")


allFeat, gt = feat2labels([feat1[:500,],feat2[:500,:]])
scaler = StandardScaler(copy=True, with_mean=True, with_std=False).fit(allFeat)
newFeat = scaler.transform(allFeat, copy=True)                           
newFeat = newFeat/np.linalg.norm(newFeat, keepdims =True, axis =1)


W, Sig, SigA, lab, c, labelK = clusterData(newFeat, numClass=5)
```
