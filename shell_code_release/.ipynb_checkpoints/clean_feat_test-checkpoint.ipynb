```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

allFeat = np.load('/tf/notebooks/clusteringTree/stl_feat/feats_2048.npy')
gtAll = np.load('/tf/notebooks/clusteringTree/stl_feat/labels_2048.npy')-1

```

```python
from sklearn.cluster import KMeans
from clusteringX2 import normIt

numClass = 2 # k-means classes

target = 0
mask = np.logical_or(gtAll==0,  gtAll ==2)
feat = allFeat[mask,:]
#gt = gtAll == target
gt = gtAll[mask]

_, m_mod = normIt(feat[::2])
feat, _ = normIt(feat, m_mod)

# feat = feat/np.linalg.norm(feat, axis =1, keepdims=True)
# m = np.mean(feat, axis =0, keepdims=True)
# feat = feat -m
# feat = feat/np.linalg.norm(feat, axis =1, keepdims=True)


kmeans = KMeans(n_clusters=numClass, random_state=0).fit(feat[::2])
labelK = kmeans.predict(feat)
means =np.zeros([numClass, feat.shape[1]])

for i in range(numClass):
    mask = labelK == i
    means[i,:] = np.mean(feat[mask,:], axis =0)

#means = means/np.linalg.norm(means, axis =1, keepdims=True)
sigInit = np.zeros([numClass])
for i in range(numClass):
    mask = np.logical_not(labelK == i)
    sigInit[i] = np.mean(np.matmul(feat[mask,:], np.transpose(means[i,:])))
```

```python
%load_ext autoreload
%autoreload
from clusteringX2 import clusterModel
from clusteringX2 import train_step2
from clusteringX2 import train_step3


tf.executing_eagerly()

model = clusterModel(means.transpose(), sigInit)
train_step3(model, feat[::2].astype("float32"), iter =100)
```

```python
c = model(feat.astype("float32"))
plt.figure(0)
plt.scatter(c[:,0], c[:,1], c= gt, alpha =0.1)

label = np.argmax(c, axis =1)
print(c.shape)
print(np.sum(np.abs(label-gt)))
print(np.sum(np.abs(labelK -gt)))
print(model.sig)
print(model.sig2)

```

```python
out_class = 4
test = allFeat[gtAll== out_class,:] 
test, _ = normIt(test, m_mod)
c_test =  model(test.astype("float32")).numpy()
plt.scatter(c_test[:,0], c_test[:,1], alpha =0.01)
plt.scatter(c[:,0], c[:,1], c=gt, alpha =0.01)
```

```python
c_ = c/np.linalg.norm(c, axis =1, keepdims=True)
c_test_ = c_test/np.linalg.norm(c_test, axis =1, keepdims=True)

plt.scatter(c_test_[:,0], c_test_[:,1], alpha =0.01)
plt.scatter(c_[:,0], c_[:,1], c=gt, alpha =0.01)
```

```python
c_orig_k = feat@kmeans.cluster_centers_.T


c_test_k =  test@kmeans.cluster_centers_.T


plt.scatter(c_test_k[:,0], c_test_k[:,1], alpha =0.1)
plt.scatter(c_orig_k[:,0], c_orig_k[:,1], c=gt, alpha =0.1)



```

```python
from sklearn.metrics import pairwise_distances

def test_feat(feat, labels, nBest= 1000):
    num_feat = feat.shape[0]
    d = pairwise_distances(feat)
    ind = np.argsort(d, axis =1)
    
    err = np.zeros(num_feat)
    
    for i in range(num_feat):
        l = labels[ind[i,:nBest+1]]      
        err[i] = nBest - (sum((l-l[0]) == 0))+1
                          
    return err
    
err = test_feat(c, gt)
print(np.mean(err))


err = test_feat(c_, gt)
print(np.mean(err))


err = test_feat(c_orig_k, gt)
print(np.mean(err))
```

```python
# with different data

gt_diff = np.concatenate([gt, -1*np.ones(c_test.shape[0])])
c_diff = np.concatenate([c_, c_test], axis =0)
c_diff_k = np.concatenate([c_orig_k, c_test_k], axis =0)


err = test_feat(c_diff, gt_diff, nBest= 1000)
print(np.mean(err))

err = test_feat(c_diff_k, gt_diff, nBest= 1000)
print(np.mean(err))

```
